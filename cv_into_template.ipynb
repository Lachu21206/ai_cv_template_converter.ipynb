```python
from google.colab import files
uploaded = files.upload()
```

```python
import re
from docx import Document

# Read uploaded file
filename = list(uploaded.keys())[0]

with open(filename, "r") as file:
    cv_text = file.read()

# Function to extract fields
def extract(pattern):
    match = re.search(pattern, cv_text)
    return match.group(1) if match else "Not Found"

name = extract(r"Name:\s*(.*)")
email = extract(r"Email:\s*(.*)")
phone = extract(r"Phone:\s*(.*)")
education = extract(r"Education:\s*(.*)")
skills = extract(r"Skills:\s*(.*)")
project = extract(r"Project:\s*(.*)")

print("Choose Template")
print("1 - Simple Template")
print("2 - Professional Template")
print("3 - Compact Template")

choice = input("Enter template number: ")

doc = Document()

if choice == "1":

    doc.add_heading("CURRICULUM VITAE", level=1)

    doc.add_paragraph(f"Name: {name}")
    doc.add_paragraph(f"Email: {email}")
    doc.add_paragraph(f"Phone: {phone}")

    doc.add_heading("Education", level=2)
    doc.add_paragraph(education)

    doc.add_heading("Skills", level=2)
    doc.add_paragraph(skills)

    doc.add_heading("Projects", level=2)
    doc.add_paragraph(project)

elif choice == "2":

    doc.add_heading("PROFESSIONAL RESUME", level=1)

    doc.add_heading("Profile", level=2)
    doc.add_paragraph(name)
    doc.add_paragraph(email)
    doc.add_paragraph(phone)

    doc.add_heading("Academic Details", level=2)
    doc.add_paragraph(education)

    doc.add_heading("Technical Skills", level=2)
    doc.add_paragraph(skills)

    doc.add_heading("Project Experience", level=2)
    doc.add_paragraph(project)

elif choice == "3":

    doc.add_heading("RESUME", level=1)

    doc.add_heading("Personal Details", level=2)
    doc.add_paragraph(name)
    doc.add_paragraph(email)
    doc.add_paragraph(phone)

    doc.add_heading("Qualification", level=2)
    doc.add_paragraph(education)

    doc.add_heading("Skills", level=2)
    doc.add_paragraph(skills)

    doc.add_heading("Projects", level=2)
    doc.add_paragraph(project)

doc.save("converted_cv.docx")

print("CV converted successfully!")
```
